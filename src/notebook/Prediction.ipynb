```python

converters = {"x":l_eval, "rho":l_eval}

df_true = pd.read_csv("../../data/df_true.csv", converters=converters)
df_pred = pd.read_csv("../../data/df_pred.csv", converters=converters)

df_true
```

```python
x = df_true.loc[0, "x"]
rho_true = df_true.loc[0, "rho"]
rho_pred = df_pred.loc[0, "rho"]
```

```python
fig, ax = plt.subplots(1,1,figsize=(6,4))
ax.plot(x, rho_true, '--', label="Label", color='tab:green', lw=1)
ax.plot(x, rho_pred, '-', label="Prédiction", color='tab:red', lw=1)
plt.ylim(-0.5, 10.1)
plt.xlabel("x")
# plt.ylabel("hauteur")
plt.legend()
plt.show()
```

```python
true = [0.875, 5.12]
pred = [0.457, 5.12]

true = np.array(true)
pred = np.array(pred)

def norm(vec):
    return vec[0]**2 + (vec[1]/10)**2

print("Erreur absolue:", round(norm(true-pred), 2))
```
